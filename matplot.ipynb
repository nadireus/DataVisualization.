```python
import seaborn as sns
```

```python
tips = sns.load_dataset ('tips')
tips.head()
```

```python
plt.scatter(tips['tip'], tips['total_bill']);
```

```python
day_of_total_bill= list(tips.groupby("day")["total_bill"].sum())
day = tips.groupby("day").sum().index   # tips.day.unique()
day
```

```python
fig, ax=plt.subplots()
ax.bar(day,day_of_total_bill )
```

```python
tip=list(tips.groupby("day")["tip"].sum())
```

```python
fig, ax=plt.subplots()
ax.barh(day,tip )
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.bar(day, day_of_total_bill,  label="total_bill")
ax.bar(day, tip,            label="tip")
plt.legend()
```

```python
# Optional hard to coding
fig, ax = plt.subplots(figsize=(10,5))
p = np.arange(len(day))
width = 0.15
ax.bar(p - width/2, day_of_total_bill, width, label="total_bill")
ax.bar(p + width/2, tip,           width, label="tip")
ax.set_xticks(p)
ax.set_xticklabels(day)
plt.legend()
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
```

```python
plt.figure(figsize=(10,8))
plt.pie(y, labels=mylabels, autopct="%.2f")
plt.show()
```

```python
plt.figure(figsize=(10,8))
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.1, 0, 0, 0]
```

```python
plt.figure(figsize=(10,8))
plt.pie(y, labels=mylabels,explode = myexplode,labeldistance=0.7, autopct="%.2f")
plt.show()
```

```python
plt.figure(figsize=(10,8))
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.1, 0.2, 0, 0]
plt.figure(figsize=(10,8))
plt.pie(y, labels=mylabels,explode = myexplode,labeldistance=0.7, autopct="%.2f")
plt.show()
```
